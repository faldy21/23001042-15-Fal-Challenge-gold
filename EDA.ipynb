# Data Cleansing

```python
alay_dict_map = dict(zip(df_kamusalay['aslinya'], df_kamusalay['pengganti']))
```

```python
alay_dict_map
```

```python
def normalize_alay(text):
    return ' '.join([alay_dict_map[word] if word in alay_dict_map else word for word in text.split(' ')])
```
